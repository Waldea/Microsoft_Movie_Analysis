```python
#Data prep
import pandas as pd
import numpy as np
```

```python
imdb_name = pd.read_csv("dsc-phase-1-project-master/zippedData/imdb.title.basics.csv.gz")

imdb_rating = pd.read_csv("dsc-phase-1-project-master/zippedData/imdb.title.ratings.csv.gz")

tn_budget = pd.read_csv('dsc-phase-1-project-master/zippedData/tn.movie_budgets.csv.gz')
```

```python
#Create function to show information on df

def information(df):
    print(df.info())
    print("NA: \n", df.isna().sum())
    print(df.duplicated().sum())
    return df.head()
```

```python
#Preparing First Dataset
information(imdb_name)
```

```python
#Remove NAs from important key columns
imdb_name = imdb_name.dropna(subset = ['original_title', 'start_year', 'genres']).reset_index(drop = 'true')

#Check for duplicates base using tconst
print(imdb_name.duplicated(subset = ['tconst']).sum())
```

```python
#Remove runtime_minutes column
imdb_name = imdb_name.drop(columns = ['runtime_minutes'], axis = 1)

information(imdb_name)
```

```python
#Preparing Second Dataset
information(imdb_rating)
```

```python
#Check for duplicates
print(imdb_rating.duplicated(subset = ['tconst']).sum())
```

```python
#Merge Dateset 1 and Dataset 2 into movies_df
movies_df = imdb_rating.merge(imdb_name, on='tconst', how='inner')
movies_df = movies_df.rename({'primary_title': 'title', 'start_year' : 'year'}, axis=1)

information(movies_df)
```

```python
print('Duplicates: ', movies_df.duplicated(subset = ['original_title']).sum())
movies_df.loc[movies_df.duplicated(subset = ['original_title', 'year'], keep=False), :]
```

```python
#Preparing Dataset 3
information(tn_budget)
```

```python
#Check for duplicates using movie title and their release date
print(tn_budget.duplicated(subset = ['movie', 'release_date']).sum())
```

```python
#Normalizing tn_budget

#Exploding release_date
tn_budget = tn_budget.assign(release_date = tn_budget['release_date'].str.split(',')).explode('release_date')

print(tn_budget.shape)
tn_budget.head(10)
```

```python
#We only want the years
#Because movies can have the same name,
#We use the movie title and production budget to identify duplicates
print(tn_budget.loc[tn_budget.duplicated(subset = ['movie', 'production_budget'], keep = False)].shape)
tn_budget.loc[tn_budget.duplicated(subset = ['movie', 'production_budget'], keep = False)].head(10)
```

```python
#Once we confirmed the shape of the dataset and the location of the rows we want for our datset,
#we can remove the unwanted rows
#keep = 'last' is the row where release date year is.
tn_budget = tn_budget.drop_duplicates(subset = ['movie', 'production_budget'], keep = 'last')

tn_budget.head()
```

```python
#Start normalizing columns
tn_budget = tn_budget.rename({'movie' : 'original_title', 'release_date' : 'year'}, axis = 1)

#Change 'year' dtype to int64
tn_budget['year'] = tn_budget['year'].astype('int64')

information(tn_budget)
```

```python
#Clean dataset 3
def clean_data(df):

    for column in ['production_budget', 'domestic_gross', 'worldwide_gross']:

        df[column] = df[column].str.replace('$','')
        df[column] = df[column].str.replace(',','')
        df[column] = df[column].astype('int64')
    return df

tn_budget_cleaned = clean_data(tn_budget)

information(tn_budget_cleaned)
```

```python
#Once we're done cleaning dataset 3,
#we can merge movie_df with tn_budget_cleaned into a master dataframe

#We need to consider movies with  the same names, hence we merge using both title and year as premise;
master_df = movies_df.merge(tn_budget_cleaned, on=['original_title', 'year'], how='inner')
information(master_df)
```

```python
#Check for duplicates using 'tconst'
master_df.duplicated('tconst').sum()
```

```python
#Tidying up master_df

#Drop unnecessary columns
master_df = master_df.drop(columns = ['tconst', 'title', 'id'])

#Rearrage columns
master_df = master_df[['original_title', 'year', 'genres', 'numvotes', 'averagerating',
                                      'production_budget', 'domestic_gross', 'worldwide_gross']]


```

```python
#Removing rows with 0 in 'production_budget', 'domestic_gross', 'worldwide_gross'
def remove_0(df):

    for column in ['production_budget', 'domestic_gross', 'worldwide_gross']:

        df = df[df[column] > 0 ]

    return df

master_df = remove_0(master_df)
master_df.shape
```

```python
#Finding dupplicates based on 'production_budget', 'domestic_gross', 'worldwide_gross'
print('Duplicates: ', master_df.duplicated(
    subset = ['original_title', 'production_budget', 'domestic_gross', 'worldwide_gross']).sum())

master_df.loc[master_df.duplicated(
    subset = ['original_title', 'production_budget', 'domestic_gross', 'worldwide_gross'], keep=False), :]
```

```python
#There are duplicate movies titles and gross but with different numvotes and average rating

#We should remove those with the lower numvotes

#Create a smaller df containing the duplicated movies with lower numvotes
df = master_df.loc[master_df.duplicated(
    subset = ['original_title', 'production_budget', 'domestic_gross', 'worldwide_gross'], keep=False), :]
df = df[df['numvotes'] < 1700]
df.shape
```

```python
#We can now remove those movies using isin function and droping na
master_df = master_df[(~master_df.isin(df))]

master_cleaned_df = master_df.dropna()

information(master_cleaned_df)
```

```python
#Final Check

def to_int(df):

    for column in ['year', 'numvotes', 'production_budget', 'domestic_gross', 'worldwide_gross']:

        df[column] = df[column].astype('int64')

    return df

master_cleaned_df = to_int(master_cleaned_df)

#Check for duplicates
duplicates =master_cleaned_df.loc[master_cleaned_df.duplicated(
    subset = ['production_budget', 'domestic_gross', 'worldwide_gross'], keep=False), :].sum()

print('Duplicates: \n', duplicates)

#Check for 0 values
num = 0
for i in master_cleaned_df.columns:
   num += master_cleaned_df[i][master_cleaned_df[i]==0].count()
print('Zero values:', num)

information(master_cleaned_df)
```

```python
#Save master_cleaned_df as CSV file
master_cleaned_df.to_csv('master_cleaned.csv', index=False)
```
